Programa que pide un número entero que este en el intervalo 18 al 25 y que siga pidiendo números mientras te mantengas en ese intervalo.


```python
n=int(input("Ingresar un numero entre el 18 y el 25: "))
while n in range(18, 26):
  print("Estas en el rango")
  n=int(input("Ingresar un numero entre el 18 y el 25: "))
else:
  print("Estas fuera de rango")
```

```python
probando = True
while probando:
  n=int(input("Ingresar un numero entre el 18 y el 25: "))
  if n in range(18, 26):
    print("Estas en el rango")
  else:
    print("Saliste de rango")
    probando = False
```

##**Dividir las cadenas en Python**<br>

Otra operación que podemos realizar a una cadena es seleccionar solamente una parte de ella. <br>Para ello se usa la notación [inicio:fin:paso] también en el nombre de la variable que almacena la cadena, donde:<br><br>
•	Inicio: es el índice del primer carácter de la porción de la cadena que queremos seleccionar.<br><br>
•	Fin: es el índice del último carácter no incluido de la porción de la cadena que queremos seleccionar.<br><br>
•	Paso: indica cada cuantos caracteres seleccionamos entre las posiciones de inicio y fin<br><br>


```python
cadena= "Programa en Python"
cadena[12:18:2]
```

##**Dar formato a las cadenas (I)**<br>
Una alternativa a la concatenación de cadenas es el uso del método format(). Este método devuelve una copia de la cadena a la que se le han sustituido las posiciones que contienen llaves ({}) por los argumentos del método. Esta sustitución se realiza por defecto en el mismo orden de los argumentos

```python
"Mi {} se llama {} y tiene {} años".format("perro", "firulais", "tres")
```

**Alternativa 2:**
También podemos referenciar los argumentos del método por su posición (siendo el índice del primero 0) o mediante un nombre.


```python
"Mi {1} se llama {2} y tiene {0} años".format("tres", "perro", "firulais")
```

##**Dividir las cadenas en Python (II)**<br>

Si omitimos las posiciones de inicio o fin, nos referimos respectivamente a la primera y última posición de la cadena. 
Además omitir el paso es equivalente a indicar que el paso es 1. De este modo las selecciones anteriores también se podrían realizar del siguiente modo.


```python
cadena= "Hola mundo"
cadena[5:]
```

##**Inmutabilidad de las cadenas python**<br>
Esto significa que una vez han sido creadas no pueden modificarse. En efecto, si intentamos modificar una cadena el intérprete nos indica que a ésta no se pueden asignar elementos.

##**Comprobar carácter en cadena<br>**
Para comprobar si un carácter está contenido en una cadena, utiliza el operador de pertenencia in. Esto nos devolverá True si al menos hay una ocurrencia del carácter en el string o False en caso contrario

```python
saludo="hola"
print('p' not in saludo)
```

##*Comprobar igualdad entre cadenas*<br>

En Python, para comparar si dos cadenas de caracteres son iguales, se utiliza el operador de igualdad ==. 
Dos strings son iguales si y solo si ambas cadenas contienen la misma secuencia de caracteres (se distingue entre mayúsculas y minúsculas).


```python
a = 'hola'
b = 'hola'
c = 'Hola'
print(a==b)
print(a==c)
```
