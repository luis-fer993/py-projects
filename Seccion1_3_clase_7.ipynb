Programa que pide un n√∫mero entero que este en el intervalo 18 al 25 y que siga pidiendo n√∫meros mientras te mantengas en ese intervalo.


```python
n=int(input("Ingresar un numero entre el 18 y el 25: "))
while n in range(18, 26):
  print("Estas en el rango")
  n=int(input("Ingresar un numero entre el 18 y el 25: "))
else:
  print("Estas fuera de rango")
```

```python
probando = True
while probando:
  n=int(input("Ingresar un numero entre el 18 y el 25: "))
  if n in range(18, 26):
    print("Estas en el rango")
  else:
    print("Saliste de rango")
    probando = False
```

##**Dividir las cadenas en Python**<br>

Otra operaci√≥n que podemos realizar a una cadena es seleccionar solamente una parte de ella. <br>Para ello se usa la notaci√≥n [inicio:fin:paso] tambi√©n en el nombre de la variable que almacena la cadena, donde:<br><br>
‚Ä¢	Inicio: es el √≠ndice del primer car√°cter de la porci√≥n de la cadena que queremos seleccionar.<br><br>
‚Ä¢	Fin: es el √≠ndice del √∫ltimo car√°cter no incluido de la porci√≥n de la cadena que queremos seleccionar.<br><br>
‚Ä¢	Paso: indica cada cuantos caracteres seleccionamos entre las posiciones de inicio y fin<br><br>


```python
cadena= "Programa en Python"
cadena[12:18:2]
```

##**Dar formato a las cadenas (I)**<br>
Una alternativa a la concatenaci√≥n de cadenas es el uso del m√©todo format(). Este m√©todo devuelve una copia de la cadena a la que se le han sustituido las posiciones que contienen llaves ({}) por los argumentos del m√©todo. Esta sustituci√≥n se realiza por defecto en el mismo orden de los argumentos

```python
"Mi {} se llama {} y tiene {} a√±os".format("perro", "firulais", "tres")
```

**Alternativa 2:**
Tambi√©n podemos referenciar los argumentos del m√©todo por su posici√≥n (siendo el √≠ndice del primero 0) o mediante un nombre.


```python
"Mi {1} se llama {2} y tiene {0} a√±os".format("tres", "perro", "firulais")
```

Adem√°s omitir el paso es equivalente a indicar que el paso es 1. De este modo las selecciones anteriores tambi√©n se podr√≠an realizar del siguiente modo.

##*Comprobar igualdad entre cadenas*<br>

En Python, para comparar si dos cadenas de caracteres son iguales, se utiliza el operador de igualdad ==. 
Dos strings son iguales si y solo si ambas cadenas contienen la misma secuencia de caracteres (se distingue entre may√∫sculas y min√∫sculas).


```python
a = 'hola'
b = 'hola'
c = 'Hola'
print(a==b)
print(a==c)
```
