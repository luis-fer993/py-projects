# Programa para repasar
Programa para repasar las capitales de Colombia.
Requisitos:
- Los departamentos han de salir en orden aleatorio.
- Si se falla se muestra la capital correcta.
- Al final del repaso vuelven a salir los fallados.
- Se seguira asi hasta que acierten todos.

```python
import random
capitales = { "Cundinamarca":"Bogota", 
             "Valle":"Cali",
             "Risaralda": "Pereira",
             "Bolivar": "Cartagena",
             "Quindio": "Armenia",
             "Cauca": "Popayan",
             "Tolima": "Ibague",
             "Boyaca": "Tunja"
             }
departamentos = list(capitales.keys())

print("-----Programa para repasar las capitales-------")
print("-----------------------------------------------")

while len(departamentos)>0:
  fallos = []  # lista vacia para guardar los fallos
  random.shuffle(departamentos) #mezclamos la lista con el metodo suffle
  for departamento in departamentos:
    print()
    print("La capital de {} es:".format(departamento))

    intento = input("--->")  #pedimos al usuario que ingrese la capital

    if intento == capitales[departamento]:   # si el valor que da el usuario es igual al valor de la clave
      print("Es correcto!")
    else:
      print("Incorrecto. La capital de {} es".format(capitales[departamento]))
      fallos.append(departamento) # a la lista de fallos se le adiciona el dep. donde se fallo
    print()
    input("'Enter' para continuar")
  departamentos=fallos # una vez se recorren la lista de los deptos. ya solo nos interesa la lista de fallos
else:
  print("")


```

#Filtrado de elementos de un array
Una caracterstica muy til de los arrays es que es muy fcil obtener otro array con los elementos que cumplen una condicin.<br>

a[condicion] : Devuelve una lista con los elementos del array a que cumplen la condicin condicion.<br>

```python
a = np.array([[1,2,3],[4,5,6]])
print(a[(a % 2 == 0)])
print(a[(a % 2 == 0) & (a>2)])
```

#Operaciones matemticas con arrays
Existen dos formas de realizar operaciones matemticas con arrays: a nivel de elemento y a nivel de array.<br>

Las operaciones a nivel de elemento operan los elementos que ocupan la misma posicin en dos arrays. Se necesitan, por tanto, dos arrays con las mismas dimensiones y el resultado es una array de la misma dimensin.<br>

Los operadores mamemticos +, -, *, /, %, ** se utilizan para la realizar suma, resta, producto, cociente, resto y potencia a nivel de elemento.<br>

```python
a = np.array([[1,2,3],[4,5,6]])
b = np.array([[1,1,1],[2,2,2]])
print(a+b)
print()
print(a/b)
print()
print(a**2)
```
